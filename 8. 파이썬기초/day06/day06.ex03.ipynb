```python
import requests

response = requests.get("https://news.naver.com/")
html = response.text
```

```python
import re

p=re.compile(r"<img[^>]+>")
img = p.findall(html)
for img in img:
    print(img)
```

```python
import re
p = re.compile(r"<span.*class=['\"]Nitem_link_menu['\"][^>]*>[^<]+</span>")
items = p.findall(html)
for item in items:
    print(item)
```
