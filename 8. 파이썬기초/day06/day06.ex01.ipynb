```python
from faker import Faker
```

```python
faker = Faker() # 기본값은 영어 데이터
faker.name()
```

```python
faker = Faker("ko-KR")
faker.name()
```

```python
members = [(faker.email(), faker.name(), faker.address())
          for i in range(0,1000)]
```

```python
for member in members:
    print(member)
# i 변수가 사용되지 않는 경우 자리만 채워준다
```
